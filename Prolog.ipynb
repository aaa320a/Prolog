#Zadanie 1

###Baza danych
```prolog
ojciec(dawid,karol).
ojciec(dawid,wiktoria).
ojciec(artur,zuzanna).
ojciec(artur,antoni).
ojciec(karol,rafal).
ojciec(karol,agnieszka).
ojciec(antoni,ksawery).
ojciec(filip,dominika).

matka(julia,karol).
matka(julia,wiktoria).
matka(laura,zuzanna).
matka(laura,antoni).
matka(zuzanna,rafal).
matka(zuzanna,agnieszka).
matka(wiktoria,dominika).
matka(zofia,ksawery).

rodzic(X,Y) :- ojciec(X,Y).
rodzic(X,Y) :- matka(X,Y).

dziecko(X,Y) :- ojciec(Y,X).
dziecko(X,Y) :- matka(Y,X).

malzenstwo(dawid,julia).
malzenstwo(artur,laura).
malzenstwo(karol,zuzanna).
malzenstwo(filip,wiktoria).
malzenstwo(antoni,zofia).

mezczyzna(dawid).
mezczyzna(karol).
mezczyzna(filip).
mezczyzna(rafal).
mezczyzna(artur).
mezczyzna(antoni).
mezczyzna(ksawery).

kobieta(julia).
kobieta(wiktoria).
kobieta(dominika).
kobieta(agnieszka).
kobieta(laura).
kobieta(zuzanna).
kobieta(zofia).

przodek(X,Y) :-
    (rodzic(X,Z), rodzic(Z,Y)) ;
    rodzic(X,Y).

brat(X,Y) :-
    rodzic(Z,X),
    rodzic(Z,Y),
    X \= Y,
    mezczyzna(X).

siostra(X,Y) :-
    rodzic(Z,X),
    rodzic(Z,Y),
    X \= Y,
    kobieta(X).

kuzyn(X,Y) :-
    rodzic(Z,X),
    rodzic(Q,Y),
    (siostra(Z,Q);
    brat(Q,Z)).
```

###Wyniki reguł
```prolog
?- przodek(karol,rafal).
true .

brat(rafal,agnieszka).
true .

?- siostra(wiktoria,karol).
true .

?- setof(X, kuzyn(X, ksawery), Wyniki).
Wyniki = [agnieszka, rafal].
```

#Zadanie 2

###Baza danych
```prolog
lotnisko(warszawa).
lotnisko(oslo).
lotnisko(lisbona).
lotnisko(praga).
lotnisko(luksemburg).

polaczenie(warszawa, oslo, 200).
polaczenie(oslo, lisbona, 250).
polaczenie(warszawa, lisbona, 400).
polaczenie(luksemburg, praga, 350).
polaczenie(praga, luksemburg, 30).
polaczenie(oslo, praga, 250).
polaczenie(luksemburg, lisbona, 150).

polaczenie(X, Y) :-
    polaczenie(X, Y, _).

koszt_podrozy(X, Y, C) :-
    polaczenie(X, Y, C).

najkrotsza_droga(X, Y, [X, Y], Calkowitykoszt) :-
    koszt_podrozy(X, Y, Calkowitykoszt).

najkrotsza_droga(X, Y, [X | Trasa], Calkowitykoszt) :-
    polaczenie(X, Z, Koszt1),
    najkrotsza_droga(Z, Y, Trasa, Koszt2),
    Calkowitykoszt is Koszt1 + Koszt2.
```

###Wyniki reguł
```prolog
?-polaczenie(warszawa,oslo).
true

?- koszt_podrozy(warszawa,lisbona,C).
C = 400 .

?- najkrotsza_droga(warszawa,praga,Trasa,Calkowitykoszt).
Trasa = [warszawa, oslo, praga],
Calkowitykoszt = 450
```

#Zadanie 3

###Baza danych
```prolog
:- dynamic produkt/3.

produkt(monitor, 10, sektor_a).
produkt(tablet, 20, sektor_a).
produkt(zelazko, 15, sektor_b).
produkt(mikrofala, 30, sektor_b).
produkt(telewizor, 20, sektor_c).
produkt(komputer, 25, sektor_c).

dostepny_produkt(X) :-
    produkt(X, Y, _),
    Y > 0.

przenies_produkt(X, Y) :-
    produkt(X, Ilosc, Skad),
    Ilosc > 0,
    retract(produkt(X, Ilosc, Skad)),
    assert(produkt(X, Ilosc, Y)),
    write(produkt(X, Ilosc, Y)).

uzupelnij_stan(X, Y) :-
    produkt(X, Ilosc, Sektor),
    Nowawartosc is Ilosc + Y,
    retract(produkt(X, Ilosc, Sektor)),
    assert(produkt(X, Nowawartosc, Sektor)),
    write(produkt(X, Nowawartosc, Y)).
```

###Wyniki reguł
```prolog
?- dostepny_produkt(monitor).
true.

?- przenies_produkt(tablet,sektor_c).
produkt(tablet,20,sektor_c)

?- uzupelnij_stan(telewizor,10).
produkt(telewizor,30,sektor_c)
```

#Zadanie 4

###Baza danych
```prolog
miasto(warszawa).
miasto(krakow).
miasto(poznan).
miasto(gdansk).
miasto(lublin).


droga(warszawa, krakow, 4).
droga(krakow, poznan, 3).
droga(poznan,gdansk, 4).
droga(gdansk,lublin, 5).
droga(lublin,warszawa, 3).

droga(X, Y) :-
    droga(X, Y, _).

czas_przejazdu(X, Y, T) :-
    droga(X, Y, T).

najkrotsza_trasa(X, Y, [X, Y], Czas) :-
    czas_przejazdu(X, Y, Czas).

najkrotsza_trasa(X, Y, [X | Trasa], Czas) :-
    droga(X, Z, Czas1),
    najkrotsza_trasa(Z, Y, Trasa, Czas2),
    Czas is Czas1 + Czas2.
```

###Wyniki reguł
```prolog
?- droga(warszawa, krakow).
true.

?- czas_przejazdu(warszawa,krakow, T).
T = 4.

?- najkrotsza_trasa(warszawa,poznan,Trasa,Czas).
Trasa = [warszawa, krakow, poznan],
Czas = 7 .
```

#Zadanie 5

###Baza danych
```prolog
ksiazka('W pustyni i w puszczy', sienkiewicz, przygodowa).
ksiazka('Hobbit', tolkien, fantasy).
ksiazka('Uczen skrytobojcy', hobb, fantasy).
ksiazka('Zbrodnia i kara', dostojewski, kryminal).
ksiazka('Kroniki Wardstone', delaney, fantasy).
ksiazka('All Tomorrows', kosemen, science_fiction).
ksiazka('Piknik na skraju drogi', strugackij, science_fiction).


ocena(uzytkownik1, 'W pustyni i w puszczy', 4).
ocena(uzytkownik1, 'Hobbit', 5).
ocena(uzytkownik1, 'Uczen skrytobojcy', 4).
ocena(uzytkownik1, 'Kroniki Wardstone', 5).
ocena(uzytkownik2, 'Zbrodnia i kara', 3).
ocena(uzytkownik2, 'All Tomorrows', 5).
ocena(uzytkownik2, 'Piknik na skraju drogi', 4).

preferencje(uzytkownik1, fantasy).
preferencje(uzytkownik2, science_fiction).

polecane_ksiazki(Uzytkownik, Gatunek) :-
    preferencje(Uzytkownik, Gatunek),
    findall(Tytul, (ksiazka(Tytul, _, Gatunek), ocena(Uzytkownik, Tytul, Ocena), Ocena >= 4), KsiaskiPolecane),
    write(KsiaskiPolecane).

dodaj_ocene(X) :-
    assertz(X).

usun_ocene(X):-
    retract(X),
    !.
```

###Wyniki reguł
```prolog
?- polecane_ksiazki(uzytkownik2, science_fiction).
[All Tomorrows,Piknik na skraju drogi]
true.
```
